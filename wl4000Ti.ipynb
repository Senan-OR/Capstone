```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import simps
```

```python
#04gs spectra at peak
data = np.loadtxt('sn2004gs_max.dat')
wavelength = data[:, 0]  # First column is wavelength
feature_flux = data[:, 1]  # Second column is flux

# Functions to normalize the flux
def normalize_flux(flux, method='min-max'):
    if method == 'min-max':
        return np.abs((flux - np.min(flux)) / (np.max(flux) - np.min(flux)))
    elif method == 'z-score':
        return np.abs((flux - np.mean(flux)) / np.std(flux))
    elif method == 'area':
        area = simps(flux, wavelength)
        return flux / area
    elif method == 'peak':
        return np.abs(flux / np.max(flux))
    elif method == 'mean':
        return np.abs(np.mean(flux))
    else:
        raise ValueError("Normalization method not recognized")

# Normalize the feature flux (choose the method you prefer)
normalized_flux = normalize_flux(feature_flux, method='peak')

# Parameters for redshift and wavelength range
redshift = 0.02
wavelength_min = 3935
wavelength_max = 4600

# Calculate rest wavelength
rest_wavelength = wavelength / (1 + redshift)

# Filter data within the specified range
mask = (rest_wavelength >= wavelength_min) & (rest_wavelength <= wavelength_max)
filtered_wavelength = rest_wavelength[mask]
filtered_flux = normalized_flux[mask]  # Use normalized flux

# Define the baseline as a linear line from the start to end of the filtered feature
start_flux, end_flux = filtered_flux[0], filtered_flux[-1]
baseline_flux = np.linspace(start_flux, end_flux, len(filtered_wavelength))

# Plot
plt.plot(filtered_wavelength, filtered_flux, label='Feature', color='blue')
plt.plot(filtered_wavelength, baseline_flux, label='Baseline', color='orange', linestyle='--')
plt.fill_between(filtered_wavelength, baseline_flux, filtered_flux, 
                 color='gray', alpha=0.5, label='Shaded Area', step='post')
plt.xlabel("Rest Wavelength")
plt.ylabel("Normalized Flux")
plt.legend()
plt.show()

# Calculate the area under the feature and baseline curves
area_feature = simps(filtered_flux, filtered_wavelength)
area_baseline = simps(baseline_flux, filtered_wavelength)

# Calculate the shaded area (difference between the feature and baseline)
shaded_area = area_baseline - area_feature

shaded_area
```
