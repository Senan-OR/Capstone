```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
SenanValues = pd.read_csv(r'C:\Users\senan\OneDrive\Desktop\Capstone\5972_pEW_CalculatedValuesMaster.csv')
UmutValues = pd.read_csv(r'C:\Users\senan\OneDrive\Desktop\Capstone\ztf_umt_subtype_5972.csv')


ztfnameSenan = np.array(SenanValues['ztfname'])
pEWSenan = np.array(SenanValues['pEW_Sil_5972_median_trap'])
pEWerrSenan = np.array(SenanValues['pEW_Sil_5972_median_trap_err'])
phaseSenan = np.array(SenanValues['phase'])

ztfnameUmut = np.array(UmutValues['ztfname'])
pEWUmut = np.array(UmutValues['pEW_Sil_5972_median_trap'])
phaseUmut = np.array(UmutValues['phase'])
pEWerrupUmut = np.array(UmutValues['pEW_err_trap_high_5972_final'])
pEWerrlowUmut = np.array(UmutValues['pEW_err_trap_low_5972_final'])
pEWerrUmut = (pEWerrupUmut + pEWerrlowUmut)/2
NewType = np.array(UmutValues['New_Type'])

pEWUmut = np.nan_to_num(pEWUmut, 0)
pEWerrUmut = np.nan_to_num(pEWerrUmut, 0)
```

```python
for i in range(len(ztfnameSenan)):
    for j in range(len(ztfnameUmut)):
        if ztfnameSenan[i] == ztfnameUmut[j]:
            name = 'grand'
            print('name grand', ztfnameSenan[i], ztfnameUmut[j])
        else:
            name = 'not grand'
```

```python
a=0
b=0
for i in range(len(ztfnameSenan)):
    for j in range(len(ztfnameUmut)):
        if ztfnameSenan[i] == ztfnameUmut[j]:
            name = 'grand'
            print(ztfnameSenan[i])
            a+=1
            #print(pEWSenan[i])
        else:
            name = 'not grand'

        if pEWUmut[j] == 0:
            pEWvalue = 'no bueno'
        else:
            pEWvalue = 'grand'

        if name == pEWvalue:
            #print(NewType[j])
            b+=1
            phase = [phaseSenan[i], phaseUmut[j]]
            pEW = [pEWSenan[i], pEWUmut[j]]
            pEWerr = [pEWerrSenan[i], pEWerrUmut[j]]
            print('phase =', phase)
            print('pEW =', pEW)
            print('pEWerr =', pEWerr)

            if NewType[j] == 'normal':
                colour = 'tab:green'
            elif NewType[j] == '91bg-like':
                colour = 'tab:red'
            elif NewType[j] == '99aa-like':
                colour = 'tab:cyan'
            elif NewType[j] == '04gs-like':
                colour = 'tab:orange'
            elif NewType[j] == '86G-like':
                colour = 'tab:brown'
            elif NewType[j] == '91T-like':
                colour = 'tab:blue'
            elif NewType[j] == 'SN Ia':
                colour = 'tab:gray'
            
            print(colour)
            plt.scatter(phase, pEW, color=colour, label = ztfnameSenan[i])
            plt.plot(phase,pEW, color=colour)
            if ztfnameSenan[i] != 'ZTF20aaymbjk':
                plt.errorbar(phase,pEW, yerr=pEWerr, color=colour)
            print("")
#plt.legend()
#plt.xlim(-9.75,-9.65)
plt.xlabel("Phase (Days from peak light)")
plt.ylabel("Pseudo equivalent width")
plt.title("Change in the Si feature near 5972$\AA$ with light curve phase")
plt.show()
print(a,b)
```
